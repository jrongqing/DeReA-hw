```python
import json

input_path = "slang/slangRL/multitask/data/source/magpie_data_source.json"
output_path = "slang/slangRL/multitask/data/source/magpie_data_source_filtered.json"

with open(input_path, "r", encoding="utf-8") as f:
    data = json.load(f)

filtered_data = [
    entry for entry in data
    if not (entry.get("usage_tag") == "figurative" and entry.get("reference_gptscore") == "literal")
]

with open(input_path, "w", encoding="utf-8") as f:
    json.dump(filtered_data, f, ensure_ascii=False, indent=2)
```

```python
import json
import random
import os


noidiom_path = "slang/detect/data/magpie_data_noidiom.json"
source_path = "slang/detect/data/magpie_data_source.json"
output_dir = "slang/detect/data/magpie_data.json"
output_path = os.path.join(output_dir, "merged_data.json")

os.makedirs(output_dir, exist_ok=True)

with open(noidiom_path, "r", encoding="utf-8") as f:
    noidiom_data = json.load(f)


sample_size = len(noidiom_data) // 2
sampled_noidiom = random.sample(noidiom_data, sample_size)


for item in sampled_noidiom:
    if "no_idiom_sentence" in item:
        item["sentence"] = item["no_idiom_sentence"]


with open(source_path, "r", encoding="utf-8") as f:
    source_data = json.load(f)


literal_data = [d for d in source_data if d.get("usage") == "literal"]


merged_data = sampled_noidiom + literal_data


with open(output_path, "w", encoding="utf-8") as f:
    json.dump(merged_data, f, ensure_ascii=False, indent=2)


```

```python
import json

input_file_path_1 = "slang/detect/data/magpie_data_source.json"
output_file_path = "slang/detect/data/magpie_data_f.json"

data = json.load(open(input_file_path_1))
data_result = []
for entry in data:
    if entry["usage"] == "figurative":
        data_result.append(entry)

print(len(data_result))
with open(output_file_path, 'w') as f:
    json.dump(data_result, f, ensure_ascii=False, indent=2)
```

```python
from tqdm import tqdm
dict_data = json.load(open("slang/detect/data/magpie_data_source.json"))
sentence_dict = []
for entry in dict_data:
    if entry["usage"] == "figurative":
        sentence_dict.append(entry["sentence"])

data = json.load(open("slang/detect/train_data/train_sft_no_idiom.json"))
data_result = []
for entry in tqdm(data):
    if entry["sentence"] in sentence_dict:
        data_result.append(entry)
print(len(data_result))
with open("slang/detect/train_data/train_sft_no_idiom_new.json", 'w') as f:
    json.dump(data_result, f, ensure_ascii=False, indent=2)
```

```python
import json
f_data = json.load(open("slang/detect/data/magpie_data_f.json"))   
w_data = json.load(open("slang/detect/data/magpie_data_noidiom.json"))


result_data = []
for entry in f_data:
    new_dict = {
        "sentence": entry["sentence"],
        "label": "1"
    }
    result_data.append(new_dict)

for entry in w_data:
    new_dict = {
        "sentence": entry["no_idiom_sentence"],
        "label": "0"
    }
    result_data.append(new_dict)

with open("slang/detect/data/magpie_data_all.json", "w") as f:
    json.dump(result_data, f, ensure_ascii=False, indent=2)
```
