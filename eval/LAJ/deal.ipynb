```python
import json
input_file_direct = "slang/eval/ende/direct/LoMI-DE-linked.json"
input_file_slangdit = "slang/eval/ende/simple_cot/simple_cot_eval.json"
input_file_cot = "slang/eval/ende/cot/Cot_eval.json"
input_file_cot_llama = "slang/eval/ende/cot/Cot_eval_llama.json"
output_file_path = "slang/eval/ende/ende_laj.json"

data_direct = json.load(open(input_file_direct))
data_slangdit = json.load(open(input_file_slangdit))
data_cot = json.load(open(input_file_cot))
data_cot_llama = json.load(open(input_file_cot_llama))

all_data = []
for entry in data_direct:
    entry["laj_label"] = "direct"
    entry["laj_translation"] = entry["direct_8b"]
    if entry["label"] == "Figurative":
        all_data.append(entry)

for entry in data_slangdit:
    entry["laj_label"] = "slangdit"
    entry["laj_translation"] = entry["simple_cot_result_translation"]
    if entry["label"] == "Figurative":
        all_data.append(entry)

for entry in data_cot:
    entry["laj_label"] = "cot"
    entry["laj_translation"] = entry["cot_translation"]
    if entry["label"] == "Figurative":
        all_data.append(entry)

for entry in data_cot_llama:
    entry["laj_label"] = "cot_llama"  
    entry["laj_translation"] = entry["cot_translation"]
    if entry["label"] == "Figurative":
        all_data.append(entry) 

with open("slang/eval/llama_detect/LAJ/laj_all.json", 'w') as f:
    json.dump(all_data, f, ensure_ascii=False, indent=2)

```
