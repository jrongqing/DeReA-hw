```python
import json
import re
input_file_path = "slang/eval/qwen_32b/LAJ/laj_simple_cot_result_translation.json"

LABEL = "simple_cot_result_translation"
input_file_path = "slang/eval/qwen_32b/LAJ/laj_"+LABEL+".json"
q_label = "q_score_"+LABEL
c_label = "c_score_"+LABEL

data = json.load(open(input_file_path))
count_q = 0
count_c = 0
all_q = 0
all_c = 0
all_count = 0
for entry in data:
    if q_label in entry:
        all_count += 1
        q = entry[q_label]
        c = entry[c_label]
        q = re.search(r'\d', q)
        c = re.search(r'\d', c)

        if q:
            all_q+=int(q.group(0))
            count_q+=1
        if c:
            all_c+=int(c.group(0))
            count_c+=1
    
avg_q = all_q/count_q
avg_c = all_c/count_c
print(all_count)
print(avg_q)
print(avg_c)

# 90.666
```
